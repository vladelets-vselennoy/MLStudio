```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=df['temp'], y=df['count'])
plt.xlabel("Temprature")
plt.ylabel("No of bicycle rented")
plt.show()
```

```python
plt.scatter(df['temp'], df['humidity'], s=df['count']/10, alpha=0.1)
plt.xlabel("Temprature")
plt.ylabel("Humidity")
plt.title("Analysing the dependancy of temp and humidity vs count")
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax.scatter(df['temp'], df['atemp'], s=df['count']/10, alpha=0.2)
a2 = fig.add_subplot(2, 1, 2)
a2.scatter(df['temp'], df['atemp'], s=df['count']/10, alpha=0.2)
```

```python

# Data (likely from a pandas dataframe) is assumed to be stored in 'df' variable
plt.scatter(x=df['windspeed'], y=df['count'], c='red', marker='.', alpha=[0.2, 0.3, 0.5, 0.7, 0.9])
plt.xlabel('windspeed')
plt.ylabel('count')
plt.title("Analysing the relationship between windspeed and the no of bicycles rented.")
plt.show()
```

```python

fig, axs = plt.subplots(1, 2)  
sizes = [x // 1000 for x in range(df.temp.count())]  
alphas = [0.1, 0.3, 0.5, 0.7, 0.9]
axs[0].scatter(df['temp'], df['humidity'], s=sizes, alpha=alphas, c='b') 

axs[0].set_xlabel('Temprature')  
axs[0].set_ylabel("Humidity")  

axs[1].scatter(df['atemp'], df['humidity'], s=sizes, alpha=alphas, c='g')  

axs[1].set_xlabel('Actual Temprature')  
axs[1].set_ylabel("Humidity")  

fig.suptitle('Temprature vs Humidity')
plt.show()
```

```python
from collections import Counter
hm = Counter(df['season'])
count = []
seas = []
for key, val in hm.items():
    count.append(val)
    seas.append(key)
print(count, seas)
# plt.pie()
```

```python
plt.pie(count, labels=seas)
plt.show()
```

```python
no_of_holidays = 0
no_of_days = 0

for val in df['holiday']:
    if val:
        no_of_holidays += 1
    no_of_days += 1

print(no_of_holidays)
plt.pie(x=[no_of_holidays, no_of_days - no_of_holidays], labels=['Holidays', 'Working days'], autopct='%1.1f%%')
plt.show()
```
